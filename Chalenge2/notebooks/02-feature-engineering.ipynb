## Cột có nhiều giá trị thiếu nhưng ý nghĩa là không có

Ví dụ các cột như: Alley, FireplaceQu, PoolQC, Fence, MiscFeature, GarageType, GarageFinish, GarageQual, GarageCond, BsmtQual, BsmtCond, BsmtExposure, BsmtFinType1, BsmtFinType2, MasVnrType

```python
import pandas as pd
from datetime import datetime
import os

# Giả sử bạn đã xử lý missing values như sau:
cols_none = [
    'Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature',
    'GarageType', 'GarageFinish', 'GarageQual', 'GarageCond',
    'BsmtQual', 'BsmtCond', 'BsmtExposure', 'BsmtFinType1', 'BsmtFinType2', 'MasVnrType'
]
# Chỉ điền giá trị thiếu cho các cột còn tồn tại
for col in cols_none:
    if col in df_train.columns:
        df_train[col] = df_train[col].fillna('None')
    if col in df_test.columns:
        df_test[col] = df_test[col].fillna('None')

# --- Ghi lại nhật ký xử lý ---
excel_file = "../reports/featureEngineering_summary.xlsx"
now = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# Kết luận
note = (
    "Đã xử lý các cột có giá trị thiếu dạng None (các cột liên quan đến thông tin Garage, Basement, "
    "và tường veneer). Sau khi xử lý, không còn giá trị NaN trong các cột này."
)

# Tạo dataframe lưu log
log_entry = pd.DataFrame({
    "Thời gian": [now],
    "Hành động": ["Xử lý Missing Values (fillna('None'))"],
    "Kết luận": [note]
})

# Nếu file chưa tồn tại, tạo mới
if not os.path.exists(excel_file):
    log_entry.to_excel(excel_file, index=False)
else:
    # Nếu đã có file, append thêm dòng mới
    old_log = pd.read_excel(excel_file)
    updated_log = pd.concat([old_log, log_entry], ignore_index=True)
    updated_log.to_excel(excel_file, index=False)

print("✅ Đã ghi log xử lý vào file Excel:", excel_file)

```

## Các cột số bị thiếu như MasVnrArea(Diện tích tường ốp gạch)

Ngta bỏ trống chắc điền 0 vào sẽ hợp lí hơn

```python
df_train['MasVnrArea'] = df_train['MasVnrArea'].fillna(0)
df_train['GarageYrBlt'] = df_train['GarageYrBlt'].fillna(0)

df_test['MasVnrArea'] = df_test['MasVnrArea'].fillna(0)
df_test['GarageYrBlt'] = df_test['GarageYrBlt'].fillna(0)
# Thêm kết luận quan trọng vào log
important_note = "Đã điền giá trị thiếu cho MasVnrArea và GarageYrBlt bằng 0. Đây là các cột số, giá trị thiếu hợp lý nhất là 0."
important_entry = pd.DataFrame({
    "Thời gian": [datetime.now().strftime("%Y-%m-%d %H:%M:%S")],
    "Hành động": ["Điền Missing Values bằng 0 cho MasVnrArea, GarageYrBlt"],
    "Kết luận": [important_note]
})

old_log = pd.read_excel(excel_file)
updated_log = pd.concat([old_log, important_entry], ignore_index=True)
updated_log.to_excel(excel_file, index=False)
print("✅ Đã ghi thêm kết luận quan trọng vào file Excel:", excel_file)
```
