## Phân tích biến mục tiêu: SalePrice

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
from datetime import datetime

# --- Tạo thư mục lưu ---
os.makedirs("../reports/figures", exist_ok=True)


# --- 1️⃣ Vẽ biểu đồ phân phối SalePrice và lưu ---
plt.figure(figsize=(8,5))
sns.histplot(df_train['SalePrice'], kde=True, bins=30, color="skyblue")
plt.title("Phân phối của SalePrice")
plt.xlabel("SalePrice")
plt.ylabel("Số lượng nhà")

figure_path = "../reports/figures/saleprice_distribution.png"
plt.savefig(figure_path, bbox_inches="tight")
plt.close()

# --- 2️⃣ Tính các chỉ số ---
mean_val = df_train['SalePrice'].mean()
median_val = df_train['SalePrice'].median()
min_val = df_train['SalePrice'].min()
max_val = df_train['SalePrice'].max()
skewness = df_train['SalePrice'].skew()

# --- 3️⃣ Lấy ngày giờ hiện tại ---
timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# --- 4️⃣ Gộp tất cả vào cùng một DataFrame ---
rows = [
    ["Ngày giờ EDA", timestamp],
    ["Nguồn dữ liệu", "data/01_raw/train.csv"],
    ["Biểu đồ lưu tại", figure_path],
    ["Mean", mean_val],
    ["Median", median_val],
    ["Min", min_val],
    ["Max", max_val],
    ["Skewness", skewness],
]

eda_summary = pd.DataFrame(rows, columns=["Thông tin", "Giá trị"])

# --- 5️⃣ Lưu vào Excel ---
excel_path = "../reports/eda_summary.xlsx"
eda_summary.to_excel(excel_path, index=False)

print(f"Đã lưu biểu đồ tại: {figure_path}")
print(f"Đã ghi EDA vào Excel: {excel_path}")
```

Nhận xét:

Phân phối lệch phải, tập trung ở giá thấp đến trung bình với đỉnh là 150000 -180000 


## 2. Ma trận tương quan giữa các biến số

```python

# --- 1️⃣ Tạo thư mục lưu ---
os.makedirs("../reports/figures", exist_ok=True)


# --- 3️⃣ Phân tích biến mục tiêu: SalePrice ---
plt.figure(figsize=(8,5))
sns.histplot(df_train['SalePrice'], kde=True, bins=30, color='skyblue')
plt.title("Phân phối của SalePrice")
plt.xlabel("SalePrice")
plt.ylabel("Số lượng nhà")

dist_path = "../reports/figures/saleprice_distribution.png"
plt.savefig(dist_path, bbox_inches="tight")
plt.close()

# --- 4️⃣ Heatmap tương quan ---
corr = df_train.corr(numeric_only=True)
plt.figure(figsize=(12,10))
sns.heatmap(corr, cmap='coolwarm', annot=False)
plt.title("Heatmap ma trận tương quan giữa các biến số")

heatmap_path = "../reports/figures/correlation_heatmap.png"
plt.savefig(heatmap_path, bbox_inches="tight")
plt.close()

# --- 5️⃣ Tính toán chỉ số thống kê ---
mean_val = df_train['SalePrice'].mean()
median_val = df_train['SalePrice'].median()
min_val = df_train['SalePrice'].min()
max_val = df_train['SalePrice'].max()
skewness = df_train['SalePrice'].skew()

# --- 6️⃣ Kết luận tự động ---
if skewness > 1:
    conclusion = "Phân phối SalePrice bị lệch phải (Right-skewed) — nên dùng log-transform."
elif skewness < -1:
    conclusion = "Phân phối SalePrice bị lệch trái (Left-skewed)."
else:
    conclusion = "Phân phối SalePrice gần như đối xứng."

# --- 7️⃣ Lấy thời gian chạy EDA ---
timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")

# --- 8️⃣ Gộp tất cả vào DataFrame chung ---
rows = [
    ["Ngày giờ EDA", timestamp],
    ["Nguồn dữ liệu", "data/01_raw/train.csv"],
    ["Biểu đồ phân phối SalePrice", dist_path],
    ["Biểu đồ heatmap tương quan", heatmap_path],
    ["Mean", mean_val],
    ["Median", median_val],
    ["Min", min_val],
    ["Max", max_val],
    ["Skewness", skewness],
    ["Kết luận", conclusion]
]

eda_summary = pd.DataFrame(rows, columns=["Thông tin", "Giá trị"])

# --- 9️⃣ Ghi vào Excel ---
excel_path = "../reports/eda_summary.xlsx"
try:
    with pd.ExcelWriter(excel_path, engine="openpyxl", mode="a" if os.path.exists(excel_path) else "w") as writer:
        eda_summary.to_excel(writer, index=False, sheet_name="EDA Summary")
    print("Đã lưu biểu đồ:")
    print(f"  - {dist_path}")
    print(f"  - {heatmap_path}")
    print(f"Đã lưu kết quả EDA vào: {excel_path}")
except PermissionError:
    print(f"Lỗi: Không thể ghi vào file '{excel_path}'. Hãy đảm bảo file này đã được đóng trong Excel hoặc các chương trình khác rồi thử lại.")
```

## Biểu đồ theo chất lượng tổng thể

```python
import openpyxl
from openpyxl.drawing.image import Image as XLImage
from openpyxl.utils import get_column_letter

# Đường dẫn file excel đã có
excel_path = "../reports/eda_summary.xlsx"

# Mở workbook và chọn sheet đầu tiên
wb = openpyxl.load_workbook(excel_path)
ws = wb.active

# Thêm ngày giờ, hình ảnh, kết luận vào các dòng tiếp theo
next_row = ws.max_row + 1

# Thêm ngày giờ
ws.cell(row=next_row, column=1, value="Ngày giờ cập nhật")
ws.cell(row=next_row, column=2, value=timestamp)

# Thêm kết luận quan trọng
ws.cell(row=next_row+1, column=1, value="Kết luận quan trọng")
ws.cell(row=next_row+1, column=2, value="Nhà có OverallQual càng cao thì SalePrice càng cao. Nên dùng biến này cho mô hình.")

# Thêm hình ảnh boxplot OverallQual
img_path = "../reports/figures/boxplot_overallqual.png"
plt.figure(figsize=(8,5))
sns.boxplot(x='OverallQual', y='SalePrice', data=df_train)
plt.title("SalePrice theo OverallQual (Chất lượng tổng thể)")
plt.savefig(img_path, bbox_inches="tight")
plt.close()

# Thêm hình ảnh vào excel (nếu openpyxl hỗ trợ)
img = XLImage(img_path)
img.anchor = f"{get_column_letter(3)}{next_row+2}"  # Đặt ảnh ở cột C, dòng tiếp theo
ws.add_image(img)

# Lưu lại file
wb.save(excel_path)
print("Đã cập nhật ngày giờ, kết luận và hình ảnh vào file Excel.")
```

Nhận xét:

Nhà có chất lượng tổng thể (OverallQual) càng cao thì giá bán (SalePrice) càng cao. Điều này hoàn toàn khớp với chỉ số tương quan 0.79 mà bạn tìm thấy trước đó. ## Biểu đồ hộp theo “BsmtQual” (Chất lượng tầng hầm)

```python
from openpyxl.drawing.image import Image as XLImage

# Vẽ và lưu boxplot BsmtQual
img_path_bsmtqual = "../reports/figures/boxplot_bsmtqual.png"
plt.figure(figsize=(8,5))
sns.boxplot(x='BsmtQual', y='SalePrice', data=df_train, order=['Fa', 'TA', 'Gd', 'Ex'])
plt.title("SalePrice theo BsmtQual (Chất lượng tầng hầm)")
plt.xlabel("BsmtQual")
plt.ylabel("SalePrice")
plt.savefig(img_path_bsmtqual, bbox_inches="tight")
plt.close()

# Mở workbook Excel

wb = openpyxl.load_workbook(excel_path)
ws = wb.active

# Tìm dòng trống tiếp theo
next_row = ws.max_row + 1

# Thêm ngày giờ
ws.cell(row=next_row, column=1, value="Ngày giờ cập nhật")
ws.cell(row=next_row, column=2, value=timestamp)

# Thêm nội dung
ws.cell(row=next_row+1, column=1, value="Nội dung")
ws.cell(row=next_row+1, column=2, value="Phân tích SalePrice theo BsmtQual (Chất lượng tầng hầm)")

# Thêm kết luận quan trọng
ws.cell(row=next_row+2, column=1, value="Kết luận quan trọng")
ws.cell(row=next_row+2, column=2, value="BsmtQual có mối quan hệ thứ tự rõ ràng với SalePrice (Fa < TA < Gd < Ex). Nên mã hóa theo thứ tự thay vì one-hot encoding.")

# Thêm đường dẫn hình ảnh
ws.cell(row=next_row+3, column=1, value="Hình ảnh")
ws.cell(row=next_row+3, column=2, value=img_path_bsmtqual)

# Thêm hình ảnh vào Excel
img_bsmtqual = XLImage(img_path_bsmtqual)
img_bsmtqual.anchor = f"C{next_row+4}"
ws.add_image(img_bsmtqual)

# Lưu file
wb.save(excel_path)
print(f"Đã thêm ngày giờ, nội dung, kết luận và hình ảnh BsmtQual vào file Excel tại dòng {next_row}.")
print(f"Hình ảnh đã lưu tại: {img_path_bsmtqual}")
```

Nhận xét:

Biểu đồ boxplot này cho thấy BsmtQual (Chất lượng tầng hầm) có mối quan hệ thứ tự (ordinal) rất rõ ràng với SalePrice. Khi chất lượng tăng lên theo thứ tự Fa (Fair/Tạm) $\rightarrow$ TA (Typical/Trung bình) $\rightarrow$ Gd (Good/Tốt) $\rightarrow$ Ex (Excellent/Xuất sắc), giá bán trung vị (đường kẻ ngang trong hộp) cũng tăng lên một cách rõ rệt.  Phân loại Ex (Xuất sắc) có giá nhà trung vị cao nhất và cũng có khoảng giá (chiều cao hộp) rộng nhất, cho thấy sự đa dạng về giá ở phân khúc cao cấp. Ngược lại, Fa có mức giá thấp và ổn định nhất. Điều này khẳng định đây là một biến phân loại có thứ tự quan trọng, và khi xử lý, nên mã hóa nó theo đúng thứ tự (ví dụ: Fa=1, TA=2, Gd=3, Ex=4) thay vì dùng one-hot.
